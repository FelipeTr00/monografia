```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Carregar o DataFrame original
#df = pd.read_csv("/home/mrs/Documents/monografia/dados/dva_wide.csv")
df = pd.read_csv("/home/mrs/Documents/monografia/dados/scatter2.csv")


# Selecionar as colunas desejadas
colunas_desejadas = ['SETOR_ECON', 'Governo', 'Pessoal', 'RCT', 'RCP']

# Criar um novo DataFrame com apenas as colunas desejadas
df_plot = df[colunas_desejadas]

# Definir o tema e a paleta de cores
sns.set_theme(style="ticks")  # Ou qualquer outro tema de sua escolha


# Criar o pairplot com a paleta de cores e o parmetro hue na parte superior

cores = {
    'Tercirio': '#2ca02c',
    'Secundrio': '#1f77b4',
    'Primrio': '#ff7f0e',
    'Sem Setor Principal': '#9467bd'
}

ax = sns.pairplot(df_plot, hue='SETOR_ECON', diag_kind='kde', palette=cores)

# Ajustar o tamanho dos marcadores
mksize = 100
for lh in ax._legend.legend_handles:  # Itera sobre os marcadores na legenda
    lh.set_alpha(1)  # Define a opacidade para evitar que os marcadores fiquem transparentes
    lh._sizes = [mksize]  # Define o tamanho dos marcadores

# Mover a legenda para a parte superior
sns.move_legend(ax, "lower center", bbox_to_anchor=(0.5, 1), ncol=5, title=None, frameon=False)

# Salvar o grfico com a legenda
ax.savefig("/home/mrs/Documents/monografia/plots/scatterplot.png")
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Carregar o DataFrame original
df = pd.read_csv("/home/mrs/Documents/monografia/dados/dva_long.csv")

# Selecionar as colunas desejadas
colunas_desejadas = ['SETOR_ECON', 'Governo', 'Pessoal']

# Criar um novo DataFrame com apenas as colunas desejadas
df_plot = df[colunas_desejadas]
# df_plot[['Governo', 'Pessoal']] = np.log(df_plot[['Governo', 'Pessoal']])

# Definir o tema e a paleta de cores
sns.set_theme(style="ticks")  # Ou qualquer outro tema de sua escolha

# Criar o pairplot com a paleta de cores e o parmetro hue na parte superior

cores = {
    'Tercirio': '#2ca02c',
    'Secundrio': '#1f77b4',
    'Primrio': '#ff7f0e',
    'Sem Setor Principal': '#9467bd'
}

ax = sns.pairplot(df_plot, hue='SETOR_ECON', diag_kind='kde', palette=cores)

# Ajustar o tamanho dos marcadores
mksize = 100
for lh in ax._legend.legendHandles:  # Itera sobre os marcadores na legenda
    lh.set_alpha(1)  # Define a opacidade para evitar que os marcadores fiquem transparentes
    lh._sizes = [mksize]  # Define o tamanho dos marcadores

# Mover a legenda para a lateral direita
sns.move_legend(ax, "center right", bbox_to_anchor=(1.0, 0.5), title=None, frameon=False)

plt.savefig("/home/mrs/Documents/monografia/plots/scatter_gov_pes.png")
```
