```python
import seaborn as sns
import pandas as pd

# Carregar o DataFrame original
df = pd.read_csv("/home/morais/Desktop/DVA_ANALISE/ANALISE1/dva_scatter2.csv")

# Selecionar as colunas desejadas
colunas_desejadas = ['GRUPO', 'Governo', 'Outros', 'Pessoal', 'RCT', 'RCP']

# Criar um novo DataFrame com apenas as colunas desejadas
df_plot = df[colunas_desejadas]

# Definir o tema e a paleta de cores
sns.set_theme(style="ticks")  # Ou qualquer outro tema de sua escolha


# Criar o pairplot com a paleta de cores e o parâmetro hue na parte superior

cores = {
    'Terciário': '#2ca02c',
    'Secundário': '#1f77b4',
    'Primário': '#ff7f0e',
    'Sem Setor Principal': '#9467bd'
}

ax = sns.pairplot(df_plot, hue='GRUPO', diag_kind='kde', palette=cores)

# Ajustar o tamanho dos marcadores
mksize = 100
for lh in ax._legend.legendHandles:  # Itera sobre os marcadores na legenda
    lh.set_alpha(1)  # Define a opacidade para evitar que os marcadores fiquem transparentes
    lh._sizes = [mksize]  # Define o tamanho dos marcadores

# Mover a legenda para a parte superior
sns.move_legend(ax, "lower center", bbox_to_anchor=(0.5, 1), ncol=5, title=None, frameon=False)


```
